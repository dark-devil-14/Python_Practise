```python
from IPython.display import display, HTML
display(HTML('<h1 style="font-size:20px; font-family: Times New Roman;">Practice Problem 1: Find if the given number is prime or not</h1>'))


#Prime number is natural number greater than 1 that has no positive divisors other than 1 and itself 
#we need to test test divisors from upto 2 to root n 
def is_prime(n):
    #Handles number less than 2 
    if(n<2):
        return False
    
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
    return True

#num = int(input("Enter Number: "))
num = 7
print(f"Is {num} a Prime Number : {is_prime(num)}")


```

```python
from IPython.display import display, HTML
display(HTML('<H1 style="font-size: 20px; font-family: Times New Roman;"> Practice Problem 2: Find the Sum of given prime number </H1>'))


#Find the sum of n prime number .
'''
find the prime number from 2 to n 
if (this_is_prime) : += i 
if(this_is_not_prime) : move to next number 
'''

def check_is_prime(n):
    if(n<2):
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            return False 
    return True
num = 98
#num = int(input("Enter the number: "))
total_sum = 0 

for i in range (2,num+1):
    if(check_is_prime(i)):
        total_sum += i

print(f"The sum of prime number till {num} is : {total_sum}")


```
