```python
#Prime number is natural number greater than 1 that has no positive divisors other than 1 and itself 
#we need to test test divisors from upto 2 to root n 

def is_prime(n):
    #Handles number less than 2 
    if(n<2):
        return False
    
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
    return True

num = int(input("Enter Number: "))
print(f"Is {num} a Prime Number : {is_prime(num)}")


```
