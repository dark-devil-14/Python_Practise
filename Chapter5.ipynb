<center> <h1> Python Loops </h1> </center>

Python has two primitive loop commands:

    while loops
    for loops
 
<b>The while Loop </b>

   1. With the while loop we can execute a set of statements as long as a condition is true.
   2. The while loop requires relevant variables to be ready, in this example we need to define an indexing variable, i, which we set to 1.

   <b> The break Statement</b>
   1. With the break statement we can stop the loop even if the while condition is true:

   <b> The continue Statement</b>
   1. With the continue statement we can stop the current iteration, and continue with the next:

   <b> The else Statement</b>
   1. With the else statement we can run a block of code once when the condition no longer is true:
   


```python
#While loop

#write a program to print from 1 to 50 
i = 0
while(i<50):
    print(i+1)
    i +=1
```

```python
#While loop example

l = ["Rajuram", "HEllo", "IITG", False, "sUBMIT", 44, 0]
i=0

while(i<len(l)):
    print(l[i])
    i +=1
```

```python
#For Loop (from 0 to n-1)
# A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

#Foor loop with LIST
fruits = ["Apple", "Bannaa", "Mango"]
for i in range(0, len(fruits)):
    print(fruits[i])

print("\n")

#for loop with tupe
tuple_1 = (4, 3, "hello", 12)
for i in range(4):
    print(tuple_1[i])

print("\n") 

#for loop with STRINGS
s = 'Hello'
for i in s:
    print(i)
```

```python
#For Loop with ELSE

l =  [1,7,8]

for iteam in l:
    print(iteam)
else:
    print("done")
```

```python
#The break Statement and Continue
#With the break statement we can stop the loop even if the while condition is true:

for i in range(100):
    if(i==4):
        break #Exit the loop right now 
    print(i)

print("\n")

for i in range(10):
    if(i==4):
        continue #Skip this iteration (no '4' is printed in the output)
    print(i)

    

```

Problem Statement 3: 

        Write a program to find that a given number is a prime number or not 

```python
# Problem Statement 3

num = int(input("Enter the given numbr: "))

for i in range(2,num):
    if(i%num) == 0:
        print(f"Sorry {num} is not a prime number.")
        break
    else:
        print(f"{num} is a PRIME number")
        break
```
