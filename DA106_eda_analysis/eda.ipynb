```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load iris dataset

df = sns.load_dataset('iris')
```

```python
df
```

```python
# Histrogram

plt.figure(figsize =(9,4))
sns.histplot(df['sepal_length'], bins=30)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
#Densitry Plot

plt.figure(figsize=(6,4))
sns.kdeplot(df['sepal_length'])
plt.title("Density Plot of Sepeal Length")
plt.show()
```

```python
#Densitry Plot

plt.figure(figsize=(6,4))
sns.kdeplot(df['sepal_length'])
sns.rugplot(df['sepal_length'])
plt.title("Density + RUG Plot of Sepeal Length")
plt.show()
```

```python
# BOX PLOT

plt.figure(figsize=(7,6))
sns.boxplot(x= 'species', y='sepal_length', data=df)
plt.title("BOX PLOT of Speal Length")
plt.show()
```

```python
#Co-relation (head_map)

plt.figure(figsize=(10,5))
sns.heatmap(df.iloc[:,0:4].corr(), annot=True, cmap='coolwarm')
plt.title("Head Map")
plt.show()
```
