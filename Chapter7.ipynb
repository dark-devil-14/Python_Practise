<H1> FILE I/O</H1>

    File handling is an important part of any web application.

    Python has several functions for creating, reading, updating, and deleting files.


    The key function for working with files in Python is the open() function.

    The open() function takes two parameters; filename, and mode.

    There are four different methods (modes) for opening a file:

    "r" - Read - Default value. Opens a file for reading, error if the file does not exist

    "a" - Append - Opens a file for appending, creates the file if it does not exist

    "w" - Write - Opens a file for writing, creates the file if it does not exist

    "x" - Create - Creates the specified file, returns an error if the file exists

    In addition you can specify if the file should be handled as binary or text mode

    "t" - Text - Default value. Text mode

    "b" - Binary - Binary mode (e.g. images)

    The game() function in program lets a user play a game and return the score as an integer. you need to read file H-SCORE.txt 
    which is either blank or contain the previous Hight score you need to write a program to pdate to Hi score whenever the game function break a Hi_score

```python
def game(score):
    try:
        with open("h-score.txt", "r") as file:
            prev_score = float(file.read())
    except FileNotFoundError:
        prev_score = 0

    if score > prev_score:
        with open("h-score.txt", "w") as file:
            file.write(str(score))
        print("ðŸŽ‰ New High Score Saved!")
    else:
        print("Score not higher than high score.")

initial_score = int(input("Score: "))
game(initial_score)
```

    Write a program to insert the table from 2 t0 20 and create a file itself

```python
def generata_table(n):
    table = ""
    for i in range(1,11):
        table += f"{n} X {i} = {n*i}\n"
    with open(f"txt_file/table_{n}.txt", "w") as f:
        f.write(table)

for i in range (2, 21):
    generata_table(i)
```
